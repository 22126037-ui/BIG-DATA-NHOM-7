```python
# Cài đặt thư viện (nếu chưa có)
!pip install kagglehub pandas -q

import kagglehub
import pandas as pd
import os

print("Đang tải toàn bộ dataset về Colab...")
# 1. Tải toàn bộ thư mục dataset (trả về đường dẫn lưu trong cache)
dataset_path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")

# 2. Tạo một Dictionary để chứa toàn bộ 9 DataFrames
dataframes = {}

# 3. Quét tất cả các file trong thư mục vừa tải
for file_name in os.listdir(dataset_path):
    if file_name.endswith('.csv'): # Chỉ lấy các file CSV
        # Bỏ đuôi '.csv' để làm tên khóa (key) cho dễ gọi
        table_name = file_name.replace('.csv', '')

        # Đọc file CSV và đưa vào dictionary
        file_path = os.path.join(dataset_path, file_name)
        dataframes[table_name] = pd.read_csv(file_path)


print("Danh sách")
for name in dataframes.keys():
    print(name)
```

```python
# Lặp qua tất cả các khóa (tên bảng) và giá trị (DataFrame) trong dictionary
for table_name, df in dataframes.items():
    print(f"BẢNG: {table_name.upper()}")
    print(f"Tổng số dòng: {len(df)}")

    # Dùng display() để bảng hiện ra đẹp hơn trong Colab
    display(df.head(5))
```

```python
import pandas as pd

# Giả sử bạn đã load các dataset vào dạng Pandas DataFrame
df_reviews = dataframes['olist_order_reviews_dataset']
df_items = dataframes['olist_order_items_dataset']
df_orders = dataframes['olist_orders_dataset']
df_customers = dataframes['olist_customers_dataset']

# 1. Gộp bảng Reviews và Items (qua order_id)
df_merged = pd.merge(df_reviews, df_items, on='order_id', how='inner')

# 2. Gộp tiếp với bảng Orders (qua order_id) để lấy customer_id
df_merged = pd.merge(df_merged, df_orders, on='order_id', how='inner')

# 3. Gộp với bảng Customers (qua customer_id) để lấy customer_unique_id
df_merged = pd.merge(df_merged, df_customers, on='customer_id', how='inner')

# 4. Tạo Rating Matrix với đúng 3 cột yêu cầu
rating_matrix = df_merged[['customer_unique_id', 'product_id', 'review_score']]

# 5. Xử lý trùng lặp (Quan trọng)
# Thư viện Surprise sẽ báo lỗi nếu 1 user đánh giá 1 item nhiều lần.
# Xử lý bằng cách lấy điểm đánh giá trung bình nếu có trùng lặp.
rating_matrix = rating_matrix.groupby(['customer_unique_id', 'product_id'])['review_score'].mean().reset_index()

print("Kích thước Rating Matrix:", rating_matrix.shape)
display(rating_matrix.head())

```

```python
!pip install scikit-surprise
from surprise import Reader, Dataset

# Khai báo thang điểm đánh giá (Rating scale) từ 1 đến 5
reader = Reader(rating_scale=(1, 5))

# Đưa DataFrame vào định dạng Dataset của Surprise
# Lưu ý: Thứ tự các cột bắt buộc phải là (User, Item, Rating)
data = Dataset.load_from_df(rating_matrix[['customer_unique_id', 'product_id', 'review_score']], reader)
```

```python
# Bước 3: Khởi tạo Model (SVD)
from surprise import SVD

# Khởi tạo thuật toán SVD
model_svd = SVD(n_factors=50,
                n_epochs=20,
                lr_all=0.005,
                reg_all=0.02,
                random_state=42) # Đặt random_state để kết quả không đổi ở mỗi lần chạy
```

```python
# Bước 4: Đánh giá mô hình (Cross Validation)
from surprise.model_selection import cross_validate

print("Đang tiến hành đánh giá mô hình (Cross-Validation cv=5)...")
# Quá trình này có thể mất 1-2 phút tùy cấu hình máy
results = cross_validate(model_svd,
                         data,
                         measures=['RMSE', 'MAE'],
                         cv=5,
                         verbose=True)
```

```python
# 1. Huấn luyện mô hình trên TOÀN BỘ dữ liệu (trước đó ta chỉ mới cắt ra để test)
trainset = data.build_full_trainset()
model_svd.fit(trainset)

# 2. Lấy ngẫu nhiên 1 khách hàng mẫu (ví dụ người đầu tiên trong danh sách)
sample_customer = rating_matrix['customer_unique_id'].iloc[0]

# 3. Lấy danh sách tất cả sản phẩm hiện có
all_products = rating_matrix['product_id'].unique()

# 4. Lấy danh sách các sản phẩm mà khách hàng này ĐÃ MUA
bought_products = rating_matrix[rating_matrix['customer_unique_id'] == sample_customer]['product_id'].unique()

# 5. Lọc ra các sản phẩm khách hàng này CHƯA MUA
unbought_products = [p for p in all_products if p not in bought_products]

# 6. Cho mô hình SVD dự đoán điểm số của khách hàng này cho các sản phẩm chưa mua
predictions = [model_svd.predict(sample_customer, p) for p in unbought_products]

# 7. Sắp xếp kết quả dự đoán từ cao xuống thấp
predictions.sort(key=lambda x: x.est, reverse=True)

# 8. In ra Top 10 Gợi ý
print(f"Khách hàng mẫu: {sample_customer}")
print("-" * 50)
print("TOP 10 SẢN PHẨM GỢI Ý (SVD):")
for i, pred in enumerate(predictions[:10]):
    print(f"{i+1}. Mã SP: {pred.iid} | Điểm dự đoán: {round(pred.est, 2)} sao")
```
